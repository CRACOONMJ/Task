```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv(r"C:\Users\omkar\OneDrive\Desktop\work\pizza sale.csv",low_memory=False)
```

```python
print(df.shape)
print(df.columns)
df.head()
```

```python
print(df.isnull().sum())
```

```python
for col in df.columns:
    if 'date' in col or 'time' in col:
        try:
            df[col] = pd.order_date(df[col], errors='coerce')
        except:
            pass

```

```python
plt.style.use('ggplot')

def plot_distribution(df, column):
    plt.figure(figsize=(10, 5))
    df[column].hist(bins=30, edgecolor='black')
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
def plot_category_counts(df, category_col):
    plt.figure(figsize=(10, 5))
    df[category_col].value_counts().plot(kind='bar')
    plt.title(f'Counts by {category_col}')
    plt.xlabel(category_col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

```

```python
plot_category_counts(df, 'pizza_category')
```

```python

import seaborn as sns
def plot_correlation_heatmap(df):
    plt.figure(figsize=(10, 8))
    corr = df.corr(numeric_only=True)
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Correlation Heatmap')
    plt.show()

```

```python
plot_correlation_heatmap(df)
```

```python
top_items = df.groupby('pizza_name_id')['quantity'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(8, 5))
sns.barplot(x=top_items.values, y=top_items.index, palette='viridis')
plt.title('Top 10 Best-Selling Pizzas')
plt.xlabel('Total Quantity Sold')
plt.tight_layout()
plt.show()

```

```python
df['Total'] = df['quantity'] * df['total_price']
print(df.columns)
df['Date'] = pd.to_datetime(df['order_date'], errors='coerce')
```

```python
df.set_index('Date', inplace=True)
monthly_sales = df['Total'].resample('M').sum()

plt.figure(figsize=(10, 5))
monthly_sales.plot()
plt.title('Monthly Pizza Sales')
plt.ylabel('Revenue')
plt.xlabel('Month')
plt.grid(True)
plt.tight_layout()
plt.show()
```
