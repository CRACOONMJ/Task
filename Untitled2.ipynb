## Import Liabaries

```python
import pandas as pd
import matplotlib.pyplot  as plt 
import seaborn as sns
```

## Load and exploring dataset

```python
titanic=pd.read_csv(r"C:\Users\omkar\OneDrive\Desktop\work\titanic.csv")
titanic.head() 
```

## exploring Data Analysics

```python
titanic.info()
```

```python
titanic.describe().T
```

```python
titanic.isnull().sum()
```

```python
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].median())
titanic.drop(columns=['Cabin'])
```

```python
average_age_by_class = titanic.groupby('Pclass')['Age'].mean()
print(average_age_by_class)
```

```python
survival_rate_by_gender = titanic.groupby('Sex')['Survived'].mean()
print(survival_rate_by_gender)
```

## Data visualize

```python
titanic['Pclass'].value_counts().plot(kind='bar')
plt.title('Number of Passengers by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
sns.pairplot(data=titanic,hue='Survived', vars=['Age', 'Fare', 'Pclass'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.histplot(titanic['Age'],bins=20,kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Pclass', y='Fare', data=titanic)
plt.title("Fare Distribution by Passenger Class")
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age',data=titanic)
plt.title("Age vs Survival")
plt.show()
```
